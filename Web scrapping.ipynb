```python
import requests
from bs4 import BeautifulSoup
import datetime
import pandas as pd
import chardet
```

```python
#urls = ["https://fr.trustpilot.com/review/orangebank.fr","https://fr.trustpilot.com/review/orangebank.fr?page=2"]
# The base URL
base_url = "https://fr.trustpilot.com/review/orangebank.fr?page="

# Initialize an empty list to store the generated URLs
urls = ["https://fr.trustpilot.com/review/orangebank.fr"]

# Loop from 2 to 164 (inclusive)
for page_number in range(2, 165):
    # Create the complete URL by appending the page number to the base URL
    url = base_url + str(page_number)
    
    # Add the URL to the list
    urls.append(url)
```

```python
for url in urls:
    response = requests.get(url)
    print(response)
```

```python
user_names= []
titles= []
ratings= []
pretty_times= []
countries= []
contents= []
for url in urls:
    
    response = requests.get(url)
    response.encoding = 'utf-8'
    html = response.content
    soup = BeautifulSoup(html, 'html.parser')
    comments = soup.find_all('article', class_="paper_paper__1PY90 paper_outline__lwsUX card_card__lQWDv card_noPadding__D8PcU styles_reviewCard__hcAvl")
    for comment in comments:
        user_name = comment.find('span',class_='typography_heading-xxs__QKBS8 typography_appearance-default__AAY17').get_text()
        country =  comment.find('div', class_= 'typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_detailsIcon__Fo_ua').get_text()
        time_string = comment.find('time').get('datetime')
        time = datetime.datetime.fromisoformat(time_string)
        pretty_time = time.strftime("%x")
        rating = comment.find('div', {'data-service-review-rating': True})['data-service-review-rating']
        title = comment.find('h2',class_="typography_heading-s__f7029 typography_appearance-default__AAY17").get_text()
        content = comment.find('p',class_="typography_body-l__KUYFJ")
        if content : 
            content = content.get_text()
        else:
            content = None
        user_names.append(user_name)
        titles.append(title)
        ratings.append(rating)
        pretty_times.append(pretty_time)
        countries.append(country)
        contents.append(content)
        
data = {
    'User Name': user_names,
    'Title': titles,
    'Rating': ratings,
    'Time': pretty_times,
    'Country': countries,
    'Content': contents,
    # Add more keys and corresponding lists for additional data
}
        
#print(titles)
df = pd.DataFrame(data)

df.to_csv('orangeScraping2.csv',index=False, encoding='utf-8-sig')
        
        
```
